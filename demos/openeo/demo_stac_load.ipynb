```python
import openeo

# Connect to the openEO backend and authenticate with EGI Check-In

connection = openeo.connect("https://openeo-dev.eodc.eu/openeo/1.1.0")
connection = connection.authenticate_oidc(provider_id="egi")
```

As outlined in the EO-MQS_DataDiscovery we can use the pystac_client in conjunction with the C-SCALE EO-MQS to easily find STAC collections to load in our openEO Process Graphs using the load_stac process.

## Step 1) Data Discovery

First we initialize the pystac_client with the eo-mqs url: https://eo-mqs.c-scale.eu/stac/v1

```python
from pystac_client import Client

client = Client.open("https://eo-mqs.c-scale.eu/stac/v1")
client.title
```

Then we get the STAC data for our desired collection, in this case the "EODC|SENTINEL2_L1C" collection.

```python
collection = client.get_collection("EODC|SENTINEL2_L1C")
collection
```

We can then select the first 10 STAC items of the collection like so:

```python
first_10 = [next(collection.get_items(recursive=True)) for _ in range(10)]

for item in first_10:
    print(item.bbox)
    print(item.datetime)
```

```python
spatial_extent={
          "west": 14.286267476378804,
          "east": 15.233660828684076,
          "south": 47.96164129726034,
          "north": 48.25103642216692
        }

temporal_extent= [
          "2020-08-01T00:00:00Z",
          "2021-08-31T00:00:00Z"
        ]


loaded_collection = connection.load_stac(url="https://eo-mqs.c-scale.eu/stac/v1/collections/EODC%7CAUSTRIA_GROUND_MOTION", spatial_extent=spatial_extent, temporal_extent=temporal_extent)
```

```python
saved_collection = loaded_collection.save_result(format="netCDF")
```

```python
job = saved_collection.create_job().start_job()
```
